## This script clones 100 projects each for Java, Python and JavaScript

Import required libraries

```python
import pandas as pd
from tqdm import tqdm
import os
```

Read the language repo file.

```python
data_frame = pd.read_csv('F:\\Dataset\\githubrepo\\langrepo.csv')
```

```python
data_frame.head()
```

Rightly assumed that respective sub-directories created for each langauge in "1-Fetching..." exists, and the projects list csv files (with the star counts) also exists.

```python
target_dir = "F:\\Dataset\\githubrepo"
```

```python
#base command for git clone
base='git clone https://github.com/'
for language in tqdm(data_frame['Language']):
    #Read each of the language files where the projects and respective star counts are
    language_data = pd.read_csv(target_dir + "\\" + language + "\\" + str(language)+'.csv')
    #Try and except blocs, in case yawa gas!
    try:
        for project in tqdm(language_data['project_name']):
            #clone each into a folder named 'repo' created where this notebook is running
            os.system(base+project+'.git repo/'+language+'/'+project.replace('/','_'))
    except:
        #Tell us the yawa if it happens
        print(os.error)
```
